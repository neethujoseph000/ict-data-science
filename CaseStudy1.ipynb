## Read the dataset to python environment


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# 1.Read the dataset to python environment

```python
df=pd.read_csv(r'C:\Users\HP\Downloads\cars_data.csv')
```

# 2.Check for null values present in the dataset

```python
df.isnull()
```

# 3.Plot a bar graph of Male v/s Female buyers participated in  sale

```python
ax=sns.countplot(x='Buyer Gender',data=df)
for p in ax.patches:
    ax.annotate('{:.0f}'.format(p.get_height()),(p.get_x()+0.25,p.get_height()+0.25))
    
plt.show()
```

# 4.Find the top 5 cars based on the sales price

```python
df.nlargest(5,['Sale Price'])
```
